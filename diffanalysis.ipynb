# Diff results

Different algorithms of subtract
--------------------------------------

We performed an Difference Image Analysis subtraction on the same images
with the ois package using *Bramich* and *AdaptiveBramich* method, and 
finally we performed the subtraction with *Properimage.propersubtract* 
method.

Analysis of subtracts
----------------------
We review the performance in terms of sources that are detectable in each 
subtract, and testing the presence of false detection using the image catalogs.
Also we are performing a series of statistical tests using pixel values.


```python
import numpy as np
from astropy.io import fits
from astropy.io import ascii
import matplotlib.pyplot as plt
import seaborn

%matplotlib inline
```

```python
ref = fits.getdata('testdiff_images/reference.fits')
new = fits.getdata('testdiff_images/aligned_newimage.fits')
diffps = fits.getdata('testdiff_images/diff.fits')
diffBr = fits.getdata('testdiff_images/diff_ois_default.fits')
diffAB = fits.getdata('testdiff_images/diff_ois_ABramich.fits')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(131)
plt.hist(diffps.flatten()/diffps.std(), log=True, label='ps', bins=15, normed=True)
plt.title('ProperS')
plt.subplot(132)
plt.hist(diffBr.flatten()/diffBr.std(), log=True, label='Br', bins=15, normed=True)
plt.title('Bramich')
plt.subplot(133)
plt.hist(diffAB.flatten()/diffAB.std(), log=True, label='AB', bins=15, normed=True)
plt.title('Adaptive B')
```
